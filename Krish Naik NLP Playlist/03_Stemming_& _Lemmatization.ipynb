In NLP all our data will be text data and we cannot give this data directly to a  machine learning model. Machine Learning model will not even understand what that text data is. 

What we do is we preprocess that the data we convert the data into some numerical represenation called **Vectors** 

**Stemming :** Process of reducing the infected words to their word stem (WiKi defination)



here from the picture we understand the common letters in the words are returned however the returned text doesnt have to meaningful.

**Lemmatizarion :** does the same thing as stemming the difference is that it converts the words and what we get back is a meaningful word. which could be understood by human beings. 



Both stemming and Lemmatization is possible using python library **NLTK** 

**Differences**

* In some cases both stemming and lemmatization give us same results. Stemming doesnt have to return meaningful words however lemmatization will return meaning words.

* Lemmatization takes more time as it has through all words and find a meaningful word. However stemming happens faster as it just has to find the base word. 

* stemming applications are sentiment analysis, postive negative sentiment anlyis and gmail spam classifier. Lemmatization can be used in chat bots, QA applications coz the reponse human gets from these application is meaningful.
